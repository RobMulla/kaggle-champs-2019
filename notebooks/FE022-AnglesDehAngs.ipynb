# Use Atom Data From Public Kernel

```python
import pandas as pd
from tqdm import tqdm
import os
%matplotlib inline
```

```python
for f in tqdm(os.listdir('../data/FE021/')):
    if 'parquet' not in f:
        continue
    f20 = f.replace('FE021','FE020')
    f22 = f.replace('FE021','FE022')
    if os.path.exists(f'../data/FE022/{f22}'):
        continue
    df = pd.read_parquet(f'../data/FE021/{f}')
    df2 = pd.read_parquet(f'../data/FE020/{f20}')
    close_atoms_cols = ['molecule_name',
                        'atom_index_0',
                        'atom_index_1',
                        'closest_to_0',
                        '2nd_closest_to_0',
                        '3rd_closest_to_0',
                        '4th_closest_to_0',
                        '5th_closest_to_0',
                        '6th_closest_to_0',
                         '7th_closest_to_0',
                         '8th_closest_to_0',
                         '9th_closest_to_0',
                         '10th_closest_to_0',
                         'closest_to_1',
                         '2nd_closest_to_1',
                         '3rd_closest_to_1',
                         '4th_closest_to_1',
                         '5th_closest_to_1',
                         '6th_closest_to_1',
                         '7th_closest_to_1',
                         '8th_closest_to_1',
                         '9th_closest_to_1',
                         '10th_closest_to_1']
    close_atoms_cols_new = [x for x in close_atoms_cols if x not in df.columns]
    df3 = pd.concat([df, df2[close_atoms_cols_new]], axis=1, sort=False)

    ang = pd.read_csv('../data/public-kernel-angles/angles.csv')
    ang2 = ang.rename(columns={'atom_index_0':'atom_index_1',
                               'atom_index_1':'atom_index_0'})
    ang_all = pd.concat([ang, ang2], sort=False)

    closest_list = ['closest_to_0',
                    '2nd_closest_to_0',
                    '3rd_closest_to_0',
                    '4th_closest_to_0',
                    '5th_closest_to_0',
                    '6th_closest_to_0',
                     '7th_closest_to_0',
                     '8th_closest_to_0',
                     '9th_closest_to_0',
                     '10th_closest_to_0',
                     'closest_to_1',
                     '2nd_closest_to_1',
                     '3rd_closest_to_1',
                     '4th_closest_to_1',
                     '5th_closest_to_1',
                     '6th_closest_to_1',
                     '7th_closest_to_1',
                     '8th_closest_to_1',
                     '9th_closest_to_1',
                     '10th_closest_to_1']
    for x in tqdm(closest_list):
        df4 = df3[['molecule_name', 'atom_index_0',x]] \
            .merge(ang_all,
                   left_on=['molecule_name', 'atom_index_0',x],
                   right_on=['molecule_name', 'atom_index_0','atom_index_1'],
                   how='left') \
            .rename(columns={'shortest_path_atoms':f'shortest_path_atoms_{x}',
                             'shortest_path_n_bonds':f'shortest_path_n_bonds_{x}',
                             'cosinus':f'cosinus_{x}',
                             'dihedral':f'dihedral_{x}'})
        df3 = pd.concat([df4[[f'shortest_path_atoms_{x}',
                                f'shortest_path_n_bonds_{x}',
                                f'cosinus_{x}',
                                f'dihedral_{x}']],
                         df3],
                       axis=1, sort=False)
    if 'id' not in df3.columns:
        df3['id'] = df2['id']
    if 'molecule_name' not in df3.columns:
        df3['molecule_name'] = df2['molecule_name']
    if 'scalar_coupling_constant' not in df3.columns:
        if 'scalar_coupling_constant' in df2.columns:
            df3['scalar_coupling_constant'] = df2['scalar_coupling_constant']
    df3 = df3.loc[:,~df3.columns.duplicated()]
    print(df.shape)
    print(df3.shape)
    df3.to_parquet(f'../data/FE022/{f22}')
```
