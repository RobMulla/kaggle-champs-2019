# Compare the Correlation Between Submissions

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
lb373 = pd.read_csv('../submissions/LB0.373_baseline_catboost_submission.csv')
M005 = pd.read_csv('../submissions/LB0.374_M005_0601_0631_submission_lgb_5folds_1.7150CV.csv')
M006 = pd.read_csv('../submissions/M006_0602_0333_submission_lgb_5folds_1.6080CV_50000iter.csv')
ss = pd.read_csv('../input/sample_submission.csv')
```

```python
oof = pd.read_csv('../oof/M006_0601_1428_oof_lgb_5folds_3.2442CV_5iter.csv')
```

```python
oof.plot(x='scalar_coupling_constant',y='oof_preds', kind='scatter', figsize=(15, 15))
```

```python
ss['lb373'] = lb373['scalar_coupling_constant']
ss['M005'] = M005['scalar_coupling_constant']
ss['M006'] = M006['scalar_coupling_constant']
```
