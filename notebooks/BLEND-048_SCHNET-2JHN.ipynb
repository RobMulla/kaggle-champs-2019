```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
sch = pd.read_csv('../scripts/hugh/inputs/data/2JHN/submit/test/submit-00010000_model-larger.csv')
b44 = pd.read_csv('../submissions/BLEND044_sub_-2.03355CV.csv')
test = pd.read_csv('../input/test.csv')
b44['sch'] = sch['scalar_coupling_constant']
b44['type'] = test['type']
```

```python
b44.loc[b44['type'] == '2JHN'].plot(x='scalar_coupling_constant',
                                    y='sch',
                                    kind='scatter')
plt.show()
```

```python
b44.loc[b44['type'] == '2JHN', 'scalar_coupling_constant'] = \
    ((b44.loc[b44['type'] == '2JHN']['sch'] * 2) + 
     (b44.loc[b44['type'] == '2JHN']['scalar_coupling_constant'] * 1)) / 3
```

```python
b44.loc[b44['type'] == '2JHN'].plot(x='scalar_coupling_constant',
                                    y='sch',
                                    kind='scatter')
plt.show()
```

```python
b44[['id','scalar_coupling_constant']] \
    .to_csv('../submissions/BLEND048_Blend44_blend_2JHN_w_schnet.csv',
            index=False)
```
