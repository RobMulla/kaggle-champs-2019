```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
sch2JHN = pd.read_csv('../scripts/hugh/inputs/data/2JHN/submit/test/submit-00010000_model-larger.csv')
sch3JHN = pd.read_csv('../scripts/hugh/inputs/data/3JHN/submit/test/submit-00040000_model-larger.csv')
b44 = pd.read_csv('../submissions/BLEND044_sub_-2.03355CV.csv')
test = pd.read_csv('../input/test.csv')
b44['scc_2jhn'] = sch2JHN['scalar_coupling_constant']
b44['scc_3jhn'] = sch3JHN['scalar_coupling_constant']
b44['type'] = test['type']
```

```python
b44.loc[b44['type'] == '2JHN'].plot(x='scalar_coupling_constant',
                                    y='scc_2jhn',
                                    kind='scatter')
plt.show()
```

```python
b44.loc[b44['type'] == '3JHN'].plot(x='scalar_coupling_constant',
                                    y='scc_3jhn',
                                    kind='scatter')
plt.show()
```

# Replace both 2JHN and 3JHN with SchNet

```python
b44.loc[b44['type'] == '2JHN', 'scalar_coupling_constant'] = b44.loc[b44['type'] == '2JHN']['scc_2jhn']
b44.loc[b44['type'] == '3JHN', 'scalar_coupling_constant'] = b44.loc[b44['type'] == '3JHN']['scc_3jhn']
```

```python
b44.loc[b44['type'] == '2JHN'].plot(x='scalar_coupling_constant',
                                    y='scc_2jhn',
                                    kind='scatter')
plt.show()
```

```python
b44.loc[b44['type'] == '3JHN'].plot(x='scalar_coupling_constant',
                                    y='scc_3jhn',
                                    kind='scatter')
plt.show()
```

```python
b44[['id','scalar_coupling_constant']] \
    .to_csv('../submissions/BLEND049_Blend044_replace_2JHN_3JHN_from_schnet.csv',
            index=False)
```
