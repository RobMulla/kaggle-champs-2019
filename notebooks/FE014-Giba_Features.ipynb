```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
tt = pd.read_csv('../data/giba_feats/train_test.csv.gz')
```

```python
tt = tt.drop(["molecule_name",
           "atom_index_0","atom_index_1",
           "type","scalar_coupling_constant",
           "ID",
           "structure_atom_0","structure_atom_1",
           "structure_x_0","structure_y_0","structure_z_0",
           "structure_x_1","structure_y_1","structure_z_1",
           "typei","pos",
           "R0","R1","E0","E1",'Unnamed: 0'], axis=1)
```

```python
tt.head()
```

```python
fe13_list = [x for x in os.listdir('../data/FE013/') if 'FE013' in x]
```

```python
for f in tqdm(fe13_list):
    print(f'running for file {f}')
    df = pd.read_parquet(f'../data/FE013/{f}')
    print('shape before merge {}'.format(df.shape))
    df = df.merge(tt, on='id', how='left')
    print('shape after merge {}'.format(df.shape))
    df.to_parquet('../data/FE014/{}'.format(f.replace('FE013','FE014')))
```
